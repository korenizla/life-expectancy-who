# Exploratory Data Analysis on Life Expectancy Dataset

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
df = pd.read_csv('../data/Life Expectancy Data.csv')
```

```python
df.info()
```

```python
# Exploring missing data
df.isnull().sum()
```

```python
# Dividing based on data types for future analysis
categorical_cols = [col for col in df.select_dtypes(include=['object']).columns]
numerical_cols = [col for col in df.select_dtypes(include=['int64', 'float64']).columns]
```

```python
# Exploring unique values in categorical data
for col in categorical_cols:
    print(f'Уникальные значения столбца {col}:', df[col].unique())
    print()
```
